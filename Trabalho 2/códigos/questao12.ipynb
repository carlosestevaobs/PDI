```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def iniciar():
    # abrindo a nova imagem
    imagem = np.array(cv2.imread('imagens/cameraman.tif', cv2.IMREAD_GRAYSCALE))

    # ajuste de brilho
    ContrasteBrilho(imagem)
    # ajuste do gamma
    CorrecaoGamma(imagem)
```

```python
def ContrasteBrilho(imagem):
    contraste = float(input("Digite o valor para o contraste (menor que 1 escurece, maior clarea)"))
    brilho = float(input("Digite o valor para o brilho (menor que 0 mais escura, maior clarea)"))

    # c.F(x,y)
    cF = cv2.multiply(imagem, contraste)
    # F(x,y) + b
    cFb = np.uint8(cv2.add(cF, brilho))
    plt.rcParams['figure.figsize'] = (16,16)
    plt.subplot(1, 2, 1)   
    plt.imshow(imagem, 'gray')
    plt.title("Imagem original")
    plt.subplot(1, 2, 2)   
    plt.imshow(cFb, 'gray')
    plt.title("Brilho = " +str(brilho)+" e contraste = "+str(contraste))
    plt.show() 
```

```python
def CorrecaoGamma(imagem):
    # recebimento do valor pelo usuário
    gamma = float(input("Digite o valor do gamma (menor que 1 realça detalhes escuros, maiores, claros)"))
    # ajuste do gamma
    imagemGamma = np.uint8(((imagem/255)**(1/gamma))*255)

    # Salvar
    plt.subplot(1, 2, 1)   
    plt.imshow(imagem, 'gray')
    plt.title("Imagem original")
    plt.subplot(1, 2, 2)   
    plt.imshow(imagemGamma, 'gray')
    plt.title("Gamma = " +str(gamma))
    plt.show()   
```

```python
iniciar()
```
