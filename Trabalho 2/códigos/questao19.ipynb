```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iniciar():
    # abrindo a imagem
    imagem = np.array(cv2.imread('imagens/linfocitos.jpg', cv2.IMREAD_GRAYSCALE))
    detectar(imagem)
```

```python
def detectar(imagem):
    # Setup SimpleBlobDetector parameters.
    params = cv2.SimpleBlobDetector_Params()

    # Change thresholds
    params.minThreshold = 0
    params.maxThreshold = 170

    # Área.
    params.filterByArea = True
    params.minArea = 100

    # Circularidade
    params.filterByCircularity = True
    params.minCircularity = 0.1

    # Convexidade
    params.filterByConvexity = True
    params.minConvexity = 0.87

    # Inércia
    params.filterByInertia = False
    params.minInertiaRatio = 0.01
    
    # Criação do detector e a passagem de parâmetros

    detector = cv2.SimpleBlobDetector_create(params)

    # detecção dos blobs.

    keypoints = detector.detect(imagem)
    
    im_with_keypoints = cv2.drawKeypoints(imagem, keypoints, np.array([]), (0,255,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    
    plt.rcParams['figure.figsize'] = (16,16)
    plt.subplot(1, 2, 1)
    plt.imshow(imagem, 'gray')
    plt.title("Corrente sanguínea")
    
    plt.subplot(1, 2, 2)   
    plt.imshow(im_with_keypoints, 'gray')
    plt.axis("off")
    cont = 0
    
    for u in keypoints:
        plt.scatter(u.pt[0],u.pt[1], color='red', s=30)
        cont = cont + 1
    plt.title("Linfócitos em destaque")
    plt.show()
    print("linfócitos encontrados: "+ format(cont))

```

```python
iniciar()
```
