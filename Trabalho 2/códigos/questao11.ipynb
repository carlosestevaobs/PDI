```python
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
def iniciar():
    # Abrindo a imagem .gif e convertendo para png
    Image.open('imagens/rat-09.gif').convert('RGB').save('imagens/rat-09.png')

    # abrindo a nova imagem
    imagem = np.array(cv2.imread('imagens/rat-09.png', cv2.IMREAD_GRAYSCALE))
    linhas, colunas = imagem.shape
    translacao(imagem, linhas, colunas)
    rotacao(imagem, linhas, colunas)
    escala(imagem, linhas, colunas)
```

```python
def translacao(imagem, linhas, colunas):
    # Translação

    # rebendo os valores de tx e ty do usuá
    print("------- Movimento de Translação -------")
    tx = int(input("Digite o valor que deseja à direita ou esquerda (valor negativo)"))
    ty = int(input("Digite o valor que deseja para cima ou baixo (valor negativo)"))

    # matriz de translação
    matTrans = np.float32([[1, 0, tx],
                           [0, 1, ty]])

    imgTrans = cv2.warpAffine(imagem, matTrans, (colunas, linhas))
    

    # Translação    
    plt.rcParams['figure.figsize'] = (16,16)
    plt.subplot(1, 2, 1)   
    plt.imshow(imagem, 'gray')
    plt.title("Imagem original")
    plt.subplot(1, 2, 2)   
    plt.imshow(imgTrans, 'gray')
    plt.title("Translação")
    plt.show()  
```

```python
def rotacao(imagem, linhas, colunas):
    # Rotação
    print("")
    print("------- Rotação -------")
    centRot = (colunas/2, linhas/2)
    graus = int(input("Digite o valor de graus a ser girado no sentido anti-horário"))

    matRot = cv2.getRotationMatrix2D(centRot, graus, 1)
    imgRot = cv2.warpAffine(imagem, matRot, (colunas, linhas))
   

    # Rotação
    plt.rcParams['figure.figsize'] = (16,16)
    plt.subplot(1, 2, 1)   
    plt.imshow(imagem, 'gray')
    plt.title("Imagem original")
    plt.subplot(1, 2, 2)
    plt.imshow(imgRot, 'gray')
    plt.title("Rotação")
    plt.show()  
```

```python
def escala(imagem, linhas, colunas):
    # Escala
    print("")
    print("------- Escala -------")

    print("Para criação da miniatura com a imagem original digite tamanhos para altura e largura entre: ")
    print(linhas)
    print(colunas)

    Nlinhas = int(input("Digite a altura do novo tamanho da imagem"))
    Ncolunas = int(input("Digite a largura"))

    plt.rcParams['figure.figsize'] = (16,16)
    imgEsc = cv2.resize(imagem, (Ncolunas, Nlinhas))
    plt.subplot(1, 2, 1)   
    plt.imshow(imagem, 'gray')
    plt.title("Imagem original")   
       
    plt.subplot(1, 2, 2)       
    plt.imshow(imgEsc, 'gray')
    plt.title("Escala")
    plt.show()      
```
