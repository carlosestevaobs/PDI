```python
import cv2
import pywt
import numpy as np
from matplotlib import pyplot as plt
```

```python
def iniciar():
    # abrindo a nova imagem
    imagem = np.array(cv2.imread('imagens/cameraman.tif', cv2.IMREAD_GRAYSCALE))
    transformadaHaar(imagem)
```

```python
def transformadaHaar(imagem):
    
    print("Tamanho original ",imagem.shape)
    imagem = pywt.dwt2(imagem, 'haar')
    cA, (cH, cV, cD) = imagem
    print("Novo tamanho ",cA.shape)
    
    # visualização   
    plt.rcParams['figure.figsize'] = (16,16)
    plt.subplot(2, 2, 1)
    plt.imshow(cA, 'gray')
    plt.title("imagem original")
    
    plt.subplot(2, 2, 2)
    plt.imshow(cH, 'gray')
    plt.title("Transformada horizontal")
    
    plt.subplot(2, 2, 3)
    plt.imshow(cV, 'gray')
    plt.title("Transformada vertical")
    
    plt.subplot(2, 2, 4)
    plt.imshow(cD, 'gray')
    plt.title("Transformada diagonal")
    
    plt.show()
```

```python
iniciar()
```
